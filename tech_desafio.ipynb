## Importando as Bibliotecas

```python
# Realizando o import das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

##Carregando as bases de dados

```python
netflix = pd.read_csv ('netflix_titles.csv')
netflix['Stream'] = 'Netflix' # Foi inserido uma coluna com o nome "Netflix", para identificar os filmes e séries da mesma
```

```python
amazon = pd.read_csv ('amazon_prime_titles.csv')
amazon['Stream'] = 'Amazon'# Foi inserido uma coluna com o nome "Amazon", para identificar os filmes e séries da mesma
```

## Mesclando as bases de dados

```python
df_streams = pd.merge(netflix, amazon, how = 'outer') # Foi utilizado a junção outer para mesclar as bases de dados, onde o index irá será utilizado caso haja necessidade, como "Primary Key"
df_streams.sort_values(by=['type','release_year'],ascending=False) 
```

## Análise exploratória e tratamento dos dados

```python
df_streams.info() # Analisando o dataset e os tipos dos dados
```

```python
# Pra uma melhor visualização das datas, elas serão separadas em 2 colunas: mes_adicionado, ano_adicionado 
df_streams['month_added'] = pd.DatetimeIndex(df_streams['date_added']).month 
df_streams['year_added'] = pd.DatetimeIndex(df_streams['date_added']).year
# Existem valores em branco nas colunas de datas, para uma melhor visualização, os valores nulos serão trocados pelos dados com maior número de repetições na coluna
df_streams['month_added'] = df_streams['month_added'].fillna(df_streams['month_added'].mode()[0])
df_streams['year_added'] = df_streams['year_added'].fillna(df_streams['year_added'].mode()[0])
df_streams.head()
```

```python
df_streams.isna().sum() # Verficando os itens nulos do dataset
```

```python
# Deletando colunas desnecessárias 
df_streams = df_streams.drop(['description', 'rating','duration','date_added','director'], axis= 1) 
df_streams.head(3)
```

```python
# Foi criado uma lista com a coluna 'cast', para que possa ser possível verificar quais são os atores que mais atuaram nos filmes
list_cast = df_streams['cast'].tolist()
list_cast
df_cast = pd.DataFrame(list_cast, columns=['cast'])
```

```python
# Após a criação do dataframe de 'atores', foi realizado um tratamento dos valores nulos, como também dos titulos que aparecem o número "1" , como atores.
df_cast = df_cast. dropna()

df_cast = df_cast.loc[(df_cast['cast'] != "1")] # Alguns filmes constava o número 1 no cast, esses valores foram retirados do dataset

# Realizei a criação de um dict para poder realizar um FOR para descobrir quais atores mais participaram dos  filmes
cast_count_dict = dict()
for index, row in df_cast.iterrows():
    names = row['cast'].split(", ")
    # print(names)
    for name in names:
      if name in cast_count_dict:
        cast_count_dict[name] = cast_count_dict[name] + 1
      else:
        cast_count_dict[name] = 1


df_cast_count = pd.DataFrame.from_dict(cast_count_dict, orient='index', columns=['total'])
```

```python
df_cast_count.sort_values(by=['total'], ascending=False)
```

```python
list_country = df_streams['country'].tolist()
list_country
df_country = pd.DataFrame(list_country, columns=['country'])
```

```python
# Após a criação do dataframe de 'country', foi realizado um tratamento dos valores nulos,
df_country = df_country. dropna()


# Realizei a criação de um dict para poder realizar um FOR para descobrir quais paises mais produziram conteúdos
country_count_dict = dict()
for index, row in df_country.iterrows():
    names = row['country'].split(", ")
    # print(names)
    for name in names:
      if name in country_count_dict:
        country_count_dict[name] = country_count_dict[name] + 1
      else:
        country_count_dict[name] = 1


country_cast_count = pd.DataFrame.from_dict(country_count_dict, orient='index', columns=['total'])
```

```python
country_cast_count.sort_values(by=['total'], ascending=False)
```

```python
# Criando algumas varáveis para gerar os gráficos necessários
freq_tv_show = df_streams[(df_streams["type"] == "TV Show")]
freq_movie = df_streams[(df_streams["type"] == "Movie")]
```

```python
df_streams['type'] = pd.Categorical(df_streams['type']) # Transformando a coluna "type" em uma categoria para melhor ser segmentada
```

```python
df_streams.head(4)
```

```python
df_streams['country'] = df_streams['country'].fillna(df_streams['country'].mode()[0])
```

```python
df_streams.head(3)
```

```python
# Criando duas colunas necessárias no dataset 
df_streams['country_principal'] = df_streams['country'].apply(lambda x: x.split(",")[0]) 
df_streams['gender_principal'] = df_streams['listed_in'].apply(lambda x: x.split(",")[0])
```

```python
list_gender = df_streams['listed_in'].tolist()
list_gender
df_gender= pd.DataFrame(list_gender, columns=['gender'])
```

```python
# Após a criação do dataframe de 'gender', foi realizado um tratamento dos valores nulos,
df_gender = df_gender. dropna()


# Criação de um dict para poder realizar um FOR para descobrir a maior quantidade dos gêneros dos filmes

gender_count_dict = dict()
for index, row in df_gender.iterrows():
    names = row['gender'].split(", ")
    # print(names)
    for name in names:
      if name in gender_count_dict:
       gender_count_dict[name] = gender_count_dict[name] + 1
      else:
        gender_count_dict[name] = 1


gender_cast_count = pd.DataFrame.from_dict(gender_count_dict, orient='index', columns=['total'])
```

```python
gender_cast_count.sort_values(by=['total'], ascending=False)
```
